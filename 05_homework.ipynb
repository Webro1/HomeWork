```python
import pandas as pd
titanic = pd.read_csv('../train.csv')
titanic
```

```python
from scipy.stats import t
import numpy as np
#1 Выживаемость всех пассажиров
std = np.std(titanic['Survived'])
alpha = 0.05
mean = titanic['Survived'].mean()
n = len(titanic['Survived'])
t.interval(1 - alpha, n - 1, loc=mean, scale=std / np.sqrt(n))
```

```python
#2 Выживаемость женщин
women_df = titanic[titanic['Sex'] == 'female']
std = np.std(women_df['Survived'])
mean = women_df['Survived'].mean()
n = len(women_df[women_df['Survived'] == 1])
t.interval(1 - alpha, n - 1, loc=mean, scale=std / np.sqrt(n))
```

```python
#3 Выживаемость людей с каютами первого класса
pclass_1 = titanic[titanic['Pclass'] == 1]
std = np.std(pclass_1['Survived'])
mean = pclass_1['Survived'].mean()
n = len(pclass_1[pclass_1['Survived'] == 1])
t.interval(1 - alpha, n - 1, loc=mean, scale=std / np.sqrt(n))
```

Normal

Проверить гипотезу о том, что женщины выживают чаще мужчин с уровнем значимости 0.05

Проверить гипотезу о том, что дети выживают чаще, чем взрослые

```python
# Проверить гипотезу о том, что женщины выживают чаще мужчин с уровнем значимости 0.05
# Гипотеза HO - что женщины выживают не чаще мужчин 
import scipy
import scipy.stats
women_df = titanic[titanic['Sex'] == 'female']
mean_women = women_df['Survived'].mean()

men_df = titanic[titanic['Sex'] == 'male']
mean_men = men_df['Survived'].mean()

alpha = 0.05
std = (np.std(women_df['Survived']) + np.std(men_df['Survived'])) / 2
n = len(titanic)

sample = np.random.normal(mean_men, std, size=n)

def t_stat(sample_mean, a0, std, df):
    return (sample_mean - a0) / (std / sqrt(df))

statistic = t_stat(sample.mean(), mean_women, sample.std(), n - 1)

critical_value = t.ppf(1 - alpha / 2, n - 1)

if abs(statistic) < critical_value:
    print('H0 справедлива')
else:
    print('H1 справедлива')
```

```python
# Гипотеза HO - что дети выживают не чаще взрослых
children_df = titanic[titanic['Age'] < 18]
mean_children = children_df['Survived'].mean()

men_df = titanic[titanic['Age'] >= 18]
mean_men = men_df['Survived'].mean()

alpha = 0.05
std = (np.std(children_df['Survived']) + np.std(men_df['Survived'])) / 2
n = len(titanic)

sample = np.random.normal(mean_men, std, size=n)

def t_stat(sample_mean, a0, std, df):
    return (sample_mean - a0) / (std / sqrt(df))

statistic = t_stat(sample.mean(), mean_children, sample.std(), n - 1)

critical_value = t.ppf(1 - alpha / 2, n - 1)

if abs(statistic) < critical_value:
    print('H0 справедлива')
else:
    print('H1 справедлива')
```
